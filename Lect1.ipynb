Question 2

```python
def fact_V(n):
    z=1
    if n==0:
        return(1)
    for i in range (1,n+1):
        z=z*i
    return(z)

print(fact_V(5))
```

Question 3

```python
def fact_V(n):
    thisdict={'Item':[],'Value':[]}
    for i in range (1,n+1):
            thisdict['Item'].append(i)
            thisdict['Value'].append(i*i)
    return (thisdict)

print(fact_V(8) )
```

Question 4

```python
def missing_char(str,i):
    s = str[:i] + str[(i+1):]
    return(s)

print(missing_char('kitten', 1))
print(missing_char('kitten', 0))
print(missing_char('kitten', 4))
    
```
